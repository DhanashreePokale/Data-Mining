# Linear Regression

It is a statistical machine learning method you can use to quantify  and make predictions based on relationships between numerical values.

2 types:
    1. Simple Linear regression
    2. Multiple Linear Regression
    
Usecases:
    1. Sales Forecasting
    2. Resource Consumption forecasting
    3. Supply cost forecasting
    4. Telecom services lifecylce forecasting
    
Assumptions:
    1. all variables are numeric, not categorical
    2. Data is free of missing values and outliers
    3. There is linear relationship between predictors and predictant
    4. all predictors are independent of each other
    5. residuals are normally distributed


```python
# Load the libraries
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb

import sklearn 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
from collections import Counter
```

```python
# Set plotting parameters for the jupyter notebook
%matplotlib inline
rcParams['figure.figsize'] = 5,4
sb.set_style('whitegrid')
```

```python
# Linear Regression on the enrollment data
address = '/Users/dhanashreepokale/Downloads/Ex_Files_Python_Data_Science_EssT/Exercise Files/Ch08/08_01/enrollment_forecast.csv'
enroll = pd.read_csv(address)

enroll.columns = ['year' , 'roll' , 'unem', 'hgrad', 'inc']
enroll.head()
```

```python
sb.pairplot(enroll)
```

```python
print(enroll.corr())
```

```python
enroll_data = enroll[['unem','hgrad']].values
enroll_target = enroll[['roll']].values
enroll_data_names = ['unem','hgrad']

X, y = scale(enroll_data), enroll_target
```

```python
# Checking missing values - Nothing is returned as there is no missing value
missing_values = X==np.NaN
X[missing_values == True]
```

```python
# Instantiate a new linear regression model and passing a parameter - Normalize = True, 
# to normalize the regression equation

LinReg = LinearRegression(normalize = True)
LinReg.fit(X,y)
```

```python
print(LinReg.score(X,y))
```
