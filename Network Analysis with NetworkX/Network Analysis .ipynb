Network
1. Body of connected data that's evaluated during graph analysis
Graph
1. A Data Visualization schematic depicting the data that comprises a network

Vocabulary
1. Nodes: the vertices around which a graph is formed
2. Edges: the lines that connect vertices within a graph structure
3. Directed graph (aka digraph): a graph where there  is a direction assigned to each edge that connects a node
4. Directed Edge: an edge feature that has been assigned a direction between nodes
5. Undirected Graph: a graph where all edges are bidirectional
6. Undirected edge: a bidirectional edge feature
7. Graph size: number of edges in a graph
8. Graph Order: number of vertices in a graph
9. Degree: number of edges connected to a vertex, with loops counted twice

Types of Graph Generators:
1. Graph Drawing Algorithms

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb

import networkx as nx
```

```python
np.set_printoptions(precision=4, suppress = True)
plt.figure(figsize =(10,3))
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

```python
# Create graph
G = nx.Graph()
nx.draw(G) # we see nothing because we created an empty graph
```

```python
G.add_node(1)
nx.draw(G)
```

```python
G.add_nodes_from([2,3,4,5,6,8,9,12,15,16])
nx.draw(G)
```

```python
# add edges
G.add_edges_from([(2,4),(2,6),(2,12),(2,8),(2,15),(3,6),(3,9),(3,12),(3,15),(4,8),(4,12),(4,16),(6,12),(8,16)])
nx.draw(G)
```

```python
# change Shapes
nx.draw_circular(G)
```

```python
# change shapes
nx.draw_spring(G)
```

```python
# labeling & coloring

nx.draw_circular(G, node_color = 'bisque', with_labels = True)
```

```python
G.remove_node(1)
nx.draw_circular(G, node_color = 'bisque', with_labels = True)
```

```python
# Identify graph properties
summary_stats = nx.info(G)
print(summary_stats)
```

```python
print(nx.degree(G))
```

```python
# the above method was lengthy
# instead we can use graph generators
G = nx.complete_graph(25)
nx.draw(G, node_color = 'bisque', with_labels = True)
```

```python
G = nx.gnc_graph(7, seed = 25)
nx.draw(G, node_color = 'bisque', with_labels = True)
```

```python
ego_G = nx.ego_graph(G, 3, radius = 5)
nx.draw(G, node_color = 'bisque', with_labels = True)
```
