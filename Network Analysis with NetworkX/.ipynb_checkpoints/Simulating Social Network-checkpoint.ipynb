# Simulating Social Network

3 easy steps:
    1. generate graph objects and edge list
    2. Assign attributes to graph nodes
    3. visualize network
    
Social Network Analysis Metrics
    1. Degree: node's connectedness
    2. Successors: a successor node is a node that could serve as a backup and potentially replace an influential node
    in a network. The function of a successor node is to preserve the flow of influence throughout a network incase
    where an important node is removed
    3. neighbors: adjacent nodes in a network
    
Considering degrees in a social network:
    In-degree: The theory  is that  if a person or profile has a lot of inbound connections, then they're considered prestigious, because many different people  and profiles want to connect with them.
    Out-degree: If a person or profile has a lot of outgoing connection, however, then they're sometimes considered influential. That's because, in theory, these people have more of a platform across which to engage and interact with their many outbound connections.
    

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb

import networkx as nx
```

```python
np.set_printoptions(precision=4, suppress = True)
plt.figure(figsize =(10,3))
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

```python
DG = nx.gn_graph(7, seed= 25)

for line in nx.generate_edgelist(DG, data = False):
    print(line)
```

```python
# Assign attributes to nodes
print(DG.node[0])
```

```python
DG.node[0]['name']= 'Alice'
print(DG.node[0])
```

```python
DG.node[1]['name'] = 'Bob'
DG.node[2]['name'] = 'Claire'
DG.node[3]['name'] = 'Denis'
DG.node[4]['name'] = 'Esther'
DG.node[5]['name'] = 'Frank'
DG.node[6]['name'] = 'George'
```

```python
DG.add_nodes_from([(0,{'age':25}),(1,{'age':31}),(2,{'age':18}),(3,{'age':47}),(4,{'age':22}),(5,{'age':23}),(6,{'age':50})])
print(DG.node[0])
```

```python
DG.node[0]['gender'] = 'f'
DG.node[1]['gender'] = 'm'
DG.node[2]['gender'] = 'f'
DG.node[3]['gender'] = 'm'
DG.node[4]['gender'] = 'f'
DG.node[5]['gender'] = 'm'
DG.node[6]['gender'] = 'm'
```

```python
nx.draw_circular(DG, node_color = 'bisque', with_labels = True)
```

```python
labeldict = {0:'Alice',1:'Bob',2:'Claire',3:'Dennis',4:'Esther',5:'Frank',6:'George'}
nx.draw_circular(DG, labels = labeldict, node_color = 'bisque', with_labels = True)
```

```python
G = DG.to_undirected()
nx.draw_spectral(G, labels = labeldict, node_color ='bisque',with_labels = True)
```

```python
# Stats on nodes and graph inspection
print(nx.info(DG))
```

```python
# considering degrees in a social network
DG.degree()
```

```python
# Identifying Successor Nodes
nx.draw_circular(DG, with_labels =True, node_color = 'bisque')
```
