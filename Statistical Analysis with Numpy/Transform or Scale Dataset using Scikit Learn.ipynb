# Transform or Scale Dataset using Scikit Learn

Essential to prevent different magnitude among the variables to cause misleading statistics and erroneous statistics.
Also, we need it for preparing data for machine learning.

There are 2 ways to scale your data.
1. Normalization - Value of observation/Sum of all observations in variable (value is between 0 & 1; relative scale)
2. Standardization - Scaling data to have 0 mean and unit variance

Scikit Learn Preprocessing Tools
1. Scale Data
2. Center Data
3. Normalize Data
4. Bin Data
5. Impute Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb

import scipy
from scipy.stats import spearmanr

import sklearn 
from sklearn import preprocessing
from sklearn.preprocessing import scale

 # this lets plots to be plotted in same window
%matplotlib inline
rcParams['figure.figsize'] = 7, 5 # 5 inch cross 4 inch figures
sb.set_style('whitegrid')
```

```python
address = '/Users/dhanashreepokale/Downloads/Ex_Files_Python_Data_Science_EssT/Exercise Files/Ch01/01_05/mtcars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
```

```python
mpg = cars['mpg']
mpg.plot()
```

```python
mpg.describe()
```

```python
# min max transformation
mpg_matrix = mpg.reshape(-1,1)
mpg_matrix
scaled = preprocessing.MinMaxScaler()
fit =scaled.fit_transform(mpg_matrix)
plt.plot(fit)
```

```python
mpg_matrix = mpg.reshape(-1,1)
mpg_matrix
scaled = preprocessing.MinMaxScaler(feature_range =(0,10))
fit =scaled.fit_transform(mpg_matrix)
plt.plot(fit)
```

```python
# Standardize
stdized_mpg = scale(mpg, axis =0, with_mean= False, with_std = False )
plt.plot(stdized_mpg)
```

```python
stdized_mpg = scale(mpg)
plt.plot(stdized_mpg) # now standardized
```
